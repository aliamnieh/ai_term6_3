```python
# TODO: تابع نمایش تصاویر رندوم را کامل کنید.

import random
from PIL import Image
def visualize_two_sample_images(directory, title="Sample Images"):
    """
    دو تصویر تصادفی از مسیر مشخص‌ شده بارگذاری و نمایش می‌دهد.

    پارامترها:
    directory (str): مسیر پوشه‌ای که تصاویر در آن قرار دارند.
    title (str): عنوانی که بالای هر تصویر نمایش داده می‌شود. (پیش‌فرض: "Sample Images")
    """
    list1=os.listdir(directory)
    subset=random.sample(list1,2)
    adresses=[]
    adresses.append(os.path.join(directory,subset[0]))
    adresses.append(os.path.join(directory,subset[1]))
    plt.figure(figsize=(10,10))
    for i in range(2):
      plt.subplot(1,2,i+1)
      img=Image.open(adresses[i])
      plt.imshow(img)
      plt.title(title)
      plt.axis('off')
    plt.show()


visualize_two_sample_images(PAPER_DIR, title="Paper")
visualize_two_sample_images(ROCK_DIR, title="Rock")
visualize_two_sample_images(SCISSORS_DIR, title="Scissors")
```

```python
# TODO: اندازه تصاویر و اندازه هر بچ (تعداد تصاویر در هر دسته آموزش) را مشخص کنید

IMG_SIZE =  (256,256)
BATCH_SIZE = 32

# TODO: داده‌های آموزش را بارگذاری کنید
import tensorflow as tf
train_ds=tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,
    subset='training',
    seed=12,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# TODO: داده‌های اعتبارسنجی را بارگذاری کنید
validation_ds=tf.keras.utils.image_dataset_from_directory(
  BASE_DIR,
  validation_split=0.2,
  subset="validation",
  seed=12,
  image_size=IMG_SIZE,
  batch_size=BATCH_SIZE)
```

```python
# TODO : کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش (فقط در صورت کافی بودن حافظه RAM)

AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.shuffle(buffer_size=2188).cache().prefetch(buffer_size=AUTOTUNE)
val_ds = validation_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# TODO: افزایش داده

import tensorflow as tf
from tensorflow.keras.layers import Input
data_augmentation = tf.keras.Sequential([
    Input(shape=(256, 256, 3)), # Add Input layer to specify input shape
    tf.keras.layers.RandomFlip('horizontal_and_vertical'),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1)
])   # TODO

plt.figure(figsize=(10,10))
for images , labels in train_ds.take(1):
  for i in range(6):
    plt.subplot(3,2,i+1)
    aug_img = data_augmentation(np.expand_dims(images[i],axis=0), training=True)
    aug_img = tf.cast(tf.clip_by_value(aug_img, 0, 255), tf.uint8).numpy()
    plt.imshow(aug_img[0])
    plt.axis('off')
plt.show()
```

```python
# TODO: نمایش داده‌های Augmented
import math
def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):
    """
    نمایش تعدادی تصویر افزایش‌ یافته (Augmented) از یک دیتاست آموزشی.

    این تابع از یک batch تصاویر موجود در دیتاست نمونه‌گیری می‌کند، لایه‌ی افزایش داده (Data Augmentation)
    را روی تصاویر اعمال کرده و سپس چند تصویر را به‌صورت شبکه‌ای در خروجی نمایش می‌دهد.

    پارامترها:
    ----------
    dataset : tf.data.Dataset
        دیتاست شامل تصاویر اولیه (مثلاً دیتاست آموزش).

    augmentation_layer : tf.keras.Sequential یا tf.keras.Layer
        لایه افزایش داده که شامل عملیات‌هایی مانند چرخش، برگرداندن یا تغییرات تصادفی دیگر است.

    num_samples : int, پیش‌فرض=9
        تعداد تصاویری که قرار است نمایش داده شوند. این عدد نباید بیشتر از تعداد تصاویر در batch باشد.

    خروجی:
    -------
    هیچ مقدار بازگردانده نمی‌شود، اما یک شکل تصویری (plot) از تصاویر افزایش‌ یافته نمایش داده می‌شود.
    """

    plt.figure(figsize=(10,10))
    for images , labels in dataset.take(1):
      batch_size=images.shape[0]
      num_samples=min(batch_size,num_samples)
      list1=random.sample(range(batch_size),num_samples)
      for i in range(num_samples):
        plt.subplot(math.ceil(num_samples/3),3,i+1)
        aug_img = augmentation_layer(np.expand_dims(images[list1[i]],axis=0), training=True)
        aug_img = tf.cast(tf.clip_by_value(aug_img, 0, 255), tf.uint8).numpy()
        plt.imshow(aug_img[0])
        plt.axis('off')
    plt.show()





visualize_augmented_images(dataset=train_ds,augmentation_layer=data_augmentation)



```

         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.

  </font>

```python
# تعریف تابع پیش‌پردازش شامل نرمال‌سازی مقادیر پیکسل
def preprocess(image, label):
    image=tf.cast(image,tf.float32)
    normalization_layer=tf.keras.layers.Rescaling(1.0/255)
    image=normalization_layer(image)
    return (image,label)

train_ds=train_ds.map(lambda x,y:(preprocess(x,y)))
validation_ds=validation_ds.map(lambda x,y:(preprocess(x,y)))
```

<br><br>
<div dir='rtl'>
  <h1><li>طراحی شبکه عصبی و اموزش مدل</li></h1>
  <font size=4>
    <li>
        در این مرحله،  باید یک مدل شبکه عصبی کانولوشنی (CNN) طراحی کنید که شامل لایه‌های استخراج ویژگی، کاهش ابعاد و لایه‌های تمام‌متصل برای طبقه‌بندی تصاویر به سه دسته مختلف باشد. همچنین، مدل باید کامپایل شده و آماده آموزش باشد.
    </li>
</font>
<br>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,BatchNormalization,Dropout
```

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

model=Sequential([

    data_augmentation,

    Conv2D(filters=32,kernel_size=(5,5),strides=(1,1),activation='relu',padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2),strides=(2,2)),

    Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),activation='relu',padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2),strides=(2,2)),

    Conv2D(filters=128,kernel_size=(3,3),strides=(1,1),activation='relu',padding='same'),
    BatchNormalization(),

    Conv2D(filters=128,kernel_size=(3,3),strides=(1,1),activation='relu',padding='same'),
    BatchNormalization(),

    Conv2D(filters=256,kernel_size=(3,3),strides=(1,1),activation='relu',padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2,2),strides=(2,2)),

    Flatten(),

    Dense(256,activation='relu'),
    Dropout(0.5),

    Dense(128,activation='relu'),
    Dropout(0.5),

    Dense(3,activation='softmax')

])
model.summary()
```

<br><br>
<div dir='rtl'>
    <font size=4>
        <li> مدل را کامپایل کنید و سپس آموزش دهید.
</div>
<br>

```python
# TODO: مدل را کامپایل کنید

from tensorflow.keras.callbacks import  EarlyStopping

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
early_stopping=EarlyStopping(
    monitor='val_accuracy',
    patience=7,
    restore_best_weights=True,
    verbose=1
)
```

<br><br>
<div dir='rtl'>
  <font size=4>
    <li>
        مدل را آموزش دهید.
</font>

```python
# TODO: مدل را آموزش دهید

history=model.fit(
    train_ds,
    epochs=50,
    batch_size=32,
    validation_data=validation_ds,
    callbacks=[early_stopping]
)
```

```python
def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.

    این تابع دو نمودار را رسم می‌کند:
    1. نمودار دقت (Accuracy) که دقت مدل در طول آموزش و اعتبارسنجی را نمایش می‌دهد.
    2. نمودار خطا (Loss) که میزان خطای مدل را در طول آموزش و اعتبارسنجی نشان می‌دهد.

    پارامترها:
    - history: شی‌ای که شامل اطلاعات تاریخچه آموزش مدل است. این شی معمولاً از متد `fit()` مدل Keras برمی‌گردد و حاوی مقادیر دقت و خطای آموزش و اعتبارسنجی برای هر اپوک است.

    نمودارهای تولید شده به صورت زیر خواهند بود:
    - در نمودار اول، دقت آموزش و اعتبارسنجی نمایش داده می‌شود.
    - در نمودار دوم، خطای آموزش و اعتبارسنجی به نمایش در می‌آید.

    خروجی:
    - نمودارهای دقت و خطا را نمایش می‌دهد.
    """
    train_losses=history.history['loss']
    val_losses=history.history['val_loss']
    train_accuracy=history.history['accuracy']
    val_accuracy=history.history['val_accuracy']

    plt.figure(figsize=(14,8))

    plt.subplot(1,2,1)
    plt.plot(train_losses,label='Training Loss')
    plt.plot(val_losses,label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(train_accuracy,label='Training Accuracy')
    plt.plot(val_accuracy,label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

plot_training_history(history)
```

        در این مرحله،  باید بتواند مدلی که آموزش داده را برای پیش‌بینی کلاس یک تصویر جدید استفاده کنید. هدف این بخش نمایش عملکرد مدل روی تصویر ورودی و نشان دادن کلاس پیش‌بینی‌شده به‌همراه میزان اطمینان مدل است. تابع زیر را تکمیل کنید تا تصویری دریافت و پیشبینی روی آن انجام دهد.

```python
from tensorflow.keras.preprocessing import image


def predict_image(model, image_path, class_names, img_size=(256, 256)):
    """
    بارگذاری و پیش‌پردازش یک تصویر و پیش‌بینی کلاس آن.

    این تابع تصویر ورودی را از مسیر مشخص‌شده بارگذاری کرده، آن را پیش‌پردازش می‌کند
    (یعنی تغییر اندازه، تبدیل به آرایه و افزودن ابعاد اضافه برای پردازش به‌صورت بچ) و سپس با استفاده از
    مدل آموزش‌دیده، پیش‌بینی انجام می‌دهد. در نهایت، کلاس پیش‌بینی‌شده و میزان اطمینان پیش‌بینی را نمایش می‌دهد.

    پارامترها:
    - model: مدل شبکه عصبی آموزش‌دیده (در اینجا مدل CNN).
    - image_path: مسیر تصویر ورودی که باید پیش‌بینی شود.
    - class_names: لیستی از نام کلاس‌ها که به مدل آموزش داده شده‌اند.
    - img_size: اندازه تصویر ورودی پس از تغییر اندازه برای مطابقت با ورودی مدل (پیش‌فرض: (180, 180)).

    خروجی:
    - predicted_class: کلاس پیش‌بینی‌شده تصویر.
    - confidence: میزان اطمینان پیش‌بینی به‌صورت درصد.

    نمودار:
    - تصویر ورودی نمایش داده می‌شود و در عنوان آن، کلاس پیش‌بینی‌شده و میزان اطمینان نمایش داده می‌شود.
    """
    img=image.load_img(image_path,target_size=img_size,color_mode='rgb')
    img_array=image.img_to_array(img)
    img_array=np.expand_dims(img_array,axis=0)/255.0
    predictions=model.predict(img_array)
    idx=np.argmax(predictions[0])
    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f'class {class_names[idx]}  confidence:{predictions[0][idx]}')
    plt.show()


# for example picture number 15 of scissors class
predict_image(model,os.path.join(SCISSORS_DIR,os.listdir(SCISSORS_DIR)[14]),['paper', 'rock', 'scissors'],img_size=(256,256))
```
